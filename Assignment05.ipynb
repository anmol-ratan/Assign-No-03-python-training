```python
# 2. Write a Python program to determine whether a given year is a leap year.

def is_leap_year(year):
  """Determines whether a given year is a leap year.

  Args:
    year: The year to check.

  Returns:
    True if the year is a leap year, False otherwise.
  """
  if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    return True
  else:
    return False

# Get input from the user
year = int(input("Enter a year: "))

# Check if it's a leap year and print the result
if is_leap_year(year):
  print(year, "is a leap year.")
else:
  print(year, "is not a leap year.")
```

```python
# 14. Write a Python program to find the date of the first Monday of a given week. Sample Year and week : 2015, 50 Expected Output : Mon Dec 14 00:00:00 2015

import datetime

def first_monday_of_week(year, week_number):
  """Finds the date of the first Monday of a given week.

  Args:
    year: The year.
    week_number: The week number.

  Returns:
    A datetime object representing the date of the first Monday of the week.
  """
  first_day_of_year = datetime.date(year, 1, 1)
  first_day_of_week = first_day_of_year + datetime.timedelta(days=(week_number - 1) * 7)

  while first_day_of_week.weekday() != 0:  # Find the first Monday
    first_day_of_week += datetime.timedelta(days=1)

  return datetime.datetime.combine(first_day_of_week, datetime.time.min)

year = 2015
week_number = 50

first_monday = first_monday_of_week(year, week_number)

print("Expected Output :", first_monday)
```

```python
# 15. Write a Python program to select all the Sundays in a specified year.

import datetime

def sundays_in_year(year):
  """Selects all the Sundays in a specified year.

  Args:
    year: The year to select Sundays from.

  Returns:
    A list of datetime objects representing the Sundays in the year.
  """
  sundays = []
  for month in range(1, 13):
    for day in range(1, 32):
      try:
        date = datetime.date(year, month, day)
        if date.weekday() == 6:  # Sunday has index 6
          sundays.append(date)
      except ValueError:
        pass  # Handle invalid dates (e.g., February 30)
  return sundays


year = 2024  # Replace with the desired year

sundays = sundays_in_year(year)

print(f"Sundays in {year}:")
for sunday in sundays:
  print(sunday)
```

```python
# 19. Write a Python program to get the date of the last Tuesday

import datetime

def last_tuesday(year, month):
  """Finds the date of the last Tuesday in a given month and year.

  Args:
    year: The year.
    month: The month.

  Returns:
    A datetime object representing the date of the last Tuesday.
  """
  first_day_of_next_month = datetime.date(year, month + 1, 1) if month < 12 else datetime.date(year + 1, 1, 1)
  last_day_of_month = first_day_of_next_month - datetime.timedelta(days=1)

  while last_day_of_month.weekday() != 1:  # Tuesday has weekday index 1
    last_day_of_month -= datetime.timedelta(days=1)

  return last_day_of_month


year = 2023
month = 10

last_tues = last_tuesday(year, month)
print(f"The last Tuesday of {month}/{year} is {last_tues}")
```

```python
#  23. Write a Python program to add a month to a specified date.

from datetime import datetime, timedelta

def add_month(date, months_to_add):
  """Adds a specified number of months to a given date.

  Args:
    date: The date to modify.
    months_to_add: The number of months to add.

  Returns:
    The updated date after adding the specified number of months.
  """
  new_year = date.year + ((date.month + months_to_add - 1) // 12)
  new_month = ((date.month + months_to_add - 1) % 12) + 1
  new_day = min(date.day, calendar.monthrange(new_year, new_month)[1])  # Adjust day if necessary

  return datetime(new_year, new_month, new_day)

# Example usage
date = datetime(2023, 10, 26)
months_to_add = 2

new_date = add_month(date, months_to_add)
print(f"Original date: {date}")
print(f"Date after adding {months_to_add} months: {new_date}")
```

```python
#  24. Write a Python program to count the number of Mondays on the 1st day of the month from 2015 to 2016.

import datetime

def count_mondays_on_first_day(start_year, end_year):
  """Counts the number of Mondays on the 1st day of the month within a range of years.

  Args:
    start_year: The starting year.
    end_year: The ending year.

  Returns:
    The count of Mondays on the 1st day of the month.
  """
  count = 0
  for year in range(start_year, end_year + 1):
    for month in range(1, 13):
      date = datetime.date(year, month, 1)
      if date.weekday() == 0:  # Monday has a weekday index of 0
        count += 1
  return count

start_year = 2015
end_year = 2016

monday_count = count_mondays_on_first_day(start_year, end_year)
print(f"Number of Mondays on the 1st day of the month from {start_year} to {end_year}: {monday_count}")
```

```python
#  27. Write a Python program to create 12 fixed dates from a specified date over a given period. The difference between two dates is 20.

import datetime

def create_fixed_dates(start_date, period, interval):
  """Creates a list of fixed dates with a specified interval from a start date.

  Args:
    start_date: The starting date (datetime.date object).
    period: The number of dates to create.
    interval: The interval between dates (in days).

  Returns:
    A list of datetime.date objects representing the fixed dates.
  """
  fixed_dates = []
  for i in range(period):
    new_date = start_date + datetime.timedelta(days=i * interval)
    fixed_dates.append(new_date)
  return fixed_dates

# Example usage:
start_date = datetime.date(2023, 10, 26)
period = 12
interval = 20

fixed_dates = create_fixed_dates(start_date, period, interval)

print("Fixed dates:")
for date in fixed_dates:
  print(date)
```

```python
# 37. Write a Python program to convert difference of two dates into days, hours, minutes, and seconds.

import datetime

def diff_dates_to_dhms(date1, date2):
  """Converts the difference between two dates into days, hours, minutes, and seconds.

  Args:
    date1: The first date (datetime.datetime object).
    date2: The second date (datetime.datetime object).

  Returns:
    A tuple containing (days, hours, minutes, seconds).
  """
  time_difference = date2 - date1
  days = time_difference.days
  seconds = time_difference.seconds
  hours = seconds // 3600
  minutes = (seconds % 3600) // 60
  seconds = seconds % 60
  return days, hours, minutes, seconds


datetime1 = datetime.datetime(2023, 10, 27, 10, 0, 0)
datetime2 = datetime.datetime(2023, 11, 10, 15, 30, 0)

days, hours, minutes, seconds = diff_dates_to_dhms(datetime1, datetime2)

print(f"The difference between {datetime1} and {datetime2} is:")
print(f"Days: {days}")
print(f"Hours: {hours}")
print(f"Minutes: {minutes}")
print(f"Seconds: {seconds}")
```

```python
#  38. Write a Python program to get the last modified information of a file.

import os
import time

def get_last_modified_info(filepath):
  """Gets the last modified information of a file.

  Args:
    filepath: The path to the file.

  Returns:
    A tuple containing the last modified timestamp (in seconds since the epoch)
    and the last modified datetime object.
  """
  try:
    modified_timestamp = os.path.getmtime(filepath)
    modified_datetime = datetime.fromtimestamp(modified_timestamp)
    return modified_timestamp, modified_datetime
  except FileNotFoundError:
    return None, None


filepath = "/tmp/gsutil_download.txt"  # Replace with the actual file path

timestamp, datetime_obj = get_last_modified_info(filepath)

if timestamp is not None:
  print(f"Last modified timestamp (seconds since epoch): {timestamp}")
  print(f"Last modified datetime: {datetime_obj}")
else:
  print(f"File not found: {filepath}")
```

```python
#  39. Write a Python program to calculate an age in years.

import datetime

def calculate_age(birth_date):
  """Calculates age in years based on a birth date.

  Args:
    birth_date: A datetime.date object representing the birth date.

  Returns:
    The age in years as an integer.
  """
  today = datetime.date.today()
  age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
  return age


birth_date = datetime.date(1990, 5, 15)  # Replace with the actual birth date

age = calculate_age(birth_date)
print(f"Age: {age} years")
```

```python
# 47. Write a Python program display a list of the dates for the 2nd Saturday of every month for a given year.

import datetime
import calendar

def second_saturday_dates(year):
  """Finds the dates of the second Saturdays of every month in a given year.

  Args:
    year: The year.

  Returns:
    A list of datetime.date objects representing the second Saturdays.
  """
  second_saturdays = []
  for month in range(1, 13):
    for day in range(1, 8):  # Check the first week for the second Saturday
      date = datetime.date(year, month, day)
      if date.weekday() == 5 and date.day >= 8:  # Saturday and day >= 8
        second_saturdays.append(date)
        break  # Found the second Saturday for this month
      elif date.weekday() == 5 and date.day < 8:
          for day in range(8, 15):
            date = datetime.date(year, month, day)
            if date.weekday() == 5:
              second_saturdays.append(date)
              break

  return second_saturdays


year = 2023  # Replace with the desired year

second_sat_list = second_saturday_dates(year)

print(f"Second Saturdays of {year}:")
for date in second_sat_list:
  print(date)
```

```python
# 50. Write a Python program to get a list of dates between two dates.

from datetime import date, timedelta


def dates_between(start_date, end_date):
  """Generates a list of dates between two dates (inclusive).

  Args:
    start_date: The starting date.
    end_date: The ending date.

  Returns:
    A list of dates between the start and end dates (inclusive).
  """
  delta = end_date - start_date  # timedelta
  days = [start_date + timedelta(days=i) for i in range(delta.days + 1)]
  return days


# Example usage:
start_date = date(2023, 10, 26)
end_date = date(2023, 10, 31)

date_list = dates_between(start_date, end_date)

print("List of dates:")
for dt in date_list:
  print(dt)
```

```python
# 52. Write a Python program to get the first and last second.

import datetime

def get_first_and_last_second_of_day(date):
  """Gets the first and last second of a given date.

  Args:
    date: The date for which to get the first and last second.

  Returns:
    A tuple containing the first and last seconds of the day as datetime objects.
  """
  first_second = datetime.datetime.combine(date, datetime.time.min)
  last_second = datetime.datetime.combine(date, datetime.time.max)
  return first_second, last_second


# Example usage:
today = datetime.date.today()
first_second, last_second = get_first_and_last_second_of_day(today)

print(f"First second of {today}: {first_second}")
print(f"Last second of {today}: {last_second}")
```

```python
# 53. Write a Python program to validate a Gregorian date. The month is between 1 and 12 inclusive, the day is within the allowed number of days for the given month. Leap year's are taken into consideration. The year is between 1 and 32767 inclusive.

import datetime

def is_valid_gregorian_date(year, month, day):
  """Validates a Gregorian date.

  Args:
    year: The year.
    month: The month (1-12).
    day: The day of the month.

  Returns:
    True if the date is valid, False otherwise.
  """
  try:
    datetime.date(year, month, day)
    return 1 <= year <= 32767 and 1 <= month <= 12 and 1 <= day <= 31
  except ValueError:
    return False

# Example usage
year = 2023
month = 12
day = 31

if is_valid_gregorian_date(year, month, day):
  print(f"{year}-{month}-{day} is a valid Gregorian date.")
else:
  print(f"{year}-{month}-{day} is not a valid Gregorian date.")
```
