 1. Write a Python program to create a class representing a stack data structure. Include methods for pushing and popping elements.



```python
class Stack:
  def __init__(self,limit=5):
    self.element = []
    self.limit = limit

  def push(self,element):
    if len(self.element)<self.limit:
      self.element.append(element)
    else:
      print("Elements can't be pushed in Stack")

  def pop(self):
    if len(self.element)>0:
      return self.element.pop()
    else:
      print("Elements can't be popped from Stack")


obj = Stack()
obj.push(7)
obj.push(6)
print(obj.element)
obj.pop()
print(obj.element)
```

2. Write a Python program to create a class representing a linked list data structure. Include methods for displaying linked list data, inserting and deleting nodes.

```python
class LinkedList:
  def __init__(self, data=None):
    self.data = data
    self.next = None

  def display(self):
    temp = self
    while temp:
      print(temp.data, end=" ")
      temp = temp.next

  def insert_at_beginning(self, new_data):
    new_node = LinkedList(new_data)
    new_node.next = self
    return new_node

  def insert_after(self, prev_node, new_data):
    if prev_node is None:
      print("The given previous node must inLinkedList.")
      return
    new_node = LinkedList(new_data)
    new_node.next = prev_node.next
    prev_node.next = new_node

  def insert_at_end(self, new_data):
    new_node = LinkedList(new_data)
    if self.data is None:
      self.data = new_data
      return
    last_node = self
    while last_node.next:
      last_node = last_node.next
    last_node.next = new_node

  def delete_node(self, key):
    if self.data == key:
      if self.next:
        self.data = self.next.data
        self.next = self.next.next
      else:
        self.data = None
      return

    temp = self
    while temp.next:
      if temp.next.data == key:
        temp.next = temp.next.next
        return
      temp = temp.next

# Example usage
llist = LinkedList(8)
llist = llist.insert_at_beginning(9)
llist = llist.insert_at_beginning(6)
llist.insert_at_end(10)
llist.insert_after(llist.next, 5)
print("Linked list:")
llist.display()
print("\nAfter deleting an element:")
llist.delete_node(8)
llist.display()






```

3. Write a Python program to create a class representing a shopping cart. Include methods for adding and removing items, and calculating the total price.

```python
class ShoppingCart:
  def __init__(self):
    self.item = []
    self.price = []


  def add_items(self,items,prices):
    self.item.append(items)
    self.price.append(prices)

  def remove_items(self,items):
    if items in self.item:
      item_index = self.item.index(items)
      self.item.remove(items)
      del self.price[item_index]


  def total_price(self):
    total_price = sum(self.price)
    return total_price

cart = ShoppingCart()
cart.add_items('apple',100)
cart.add_items('banana',50)
cart.remove_items('apple')
print(cart.total_price())
```

4. Write a Python program to create a class representing a stack data structure. Include methods for pushing, popping and displaying elements.

```python
class Stack:
  def __init__(self,limit=5):
    self.elements = []
    self.limit = limit

  def push(self,element):
    if len(self.elements)<self.limit:
      self.elements.append(element)
    else:
      print("Elements can't be pushed in Stack")

  def pop(self):
    if len(self.elements)>0:
      return self.elements.pop()
    else:
      print("Elements can't be popped from Stack")

  def display(self):
    for element in self.elements:
      print(element)



obj = Stack()
obj.push(7)
obj.push(6)
print(obj.elements)
obj.pop()
print(obj.elements)
obj.display()
```
