```python
import re
from cryptography.fernet import Fernet

def get_reference_id():
    ref_id = input("Enter the 12-character Reference ID (alphanumeric): ")
    return ref_id

def validate_reference_id(ref_id):
    if len(ref_id) != 12 or not re.match("^[A-Za-z0-9]*$", ref_id):
        raise ValueError("Reference ID must be 12 alphanumeric characters.")
    return True

def encrypt_reference_id(ref_id, key):
    cipher = Fernet(key)
    encrypted_ref_id = cipher.encrypt(ref_id.encode())
    return encrypted_ref_id

key = Fernet.generate_key()
cipher_suite = Fernet(key)

try:
    reference_id = get_reference_id()
    if validate_reference_id(reference_id):
        encrypted_id = encrypt_reference_id(reference_id, key)
        print(f"Encrypted Reference ID: {encrypted_id.decode()}")

except ValueError as e:
    print(e)
```
